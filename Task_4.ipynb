```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop("id", axis = 1)
```

```python
df = df.drop("Unnamed: 32", axis = 1)
```

```python
df.describe().transpose()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["diagnosis"] = le.fit_transform(df["diagnosis"])
```

```python
df.head()
```

```python
sns.countplot(data = df, x = "diagnosis")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("diagnosis", axis = 1)
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
proba = model.predict_proba(X_test)[:, 1]
```

```python
proba
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
accuracy_score(y_test, predictions)
```

```python
precision_score(y_test, predictions)
```

```python
recall_score(y_test, predictions)
```

```python
f1_score(y_test, predictions)
```

```python
roc_auc_score(y_test, predictions)
```

```python
sns.heatmap(confusion_matrix(y_test, predictions), annot = True, cmap = "viridis")
plt.title("Confusion Matrix")
plt.xlabel("predicted")
plt.ylabel("actual")
```

```python
fpr, tpr, thresholds = roc_curve(y_test, proba)
plt.figure()
plt.plot(fpr, tpr, label = "Logistic Regression")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.title("ROC Curve")
plt.legend()
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, proba)
```

```python
plt.plot(thresholds, precision[:-1], label = "Precision")
plt.plot(thresholds, recall[:-1], label = "Recall")
plt.xlabel("threshold")
plt.ylabel("score")
plt.title('Precision vs Recall for Different Thresholds')
plt.legend()
plt.grid()
```

Logistic Regression uses the sigmoid function to convert input values (z) into a probability between 0 and 1:

\[
\sigma(z) = \frac{1}{1 + e^{-z}}
\]

- If the output is > 0.5 → class 1
- If the output is ≤ 0.5 → class 0

This probability threshold can be changed (tuned) to improve model performance depending on the problem — e.g., for imbalanced data.
